```python
# All column info
def get_dataframe_data(df):
    # Create Dataframe on all column info
    all_data = pd.DataFrame(
    {'columns': df.columns,
     'type': df.dtypes,
     'count_of_nans': df.isnull().sum()
    }).set_index('columns')
    
    # Booleon if column has NaN's
    all_data['no_nans'] = all_data['count_of_nans'].map(lambda x: x > 0)
    
    df_len = len(df.index)
    
    return all_data, df_len
```

```python
def null_removel(df):
    # Drop row if any value is NaN
    df.dropna()
    
    # Drop row only if all columns are NaN
    df.dropna(how='all')
    
    # Drop row if NaN within specific column name
    df.dropna(subset = ['column_name'])


def imputation(df):
    # Fill NaN's with the mean of the column
    df['price'].fillna(df['price'].mean(), inplace=True)
    
    # Fill NaN's with the medium of the column
    df['price'].fillna(df['price'].medium(), inplace=True)
```
